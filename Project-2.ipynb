1. Напишите запрос, который позволит узнать, сколько (cnt) вакансий в каждом регионе (area).
Отсортируйте по количеству вакансий в порядке убывания.

```python
# текст запроса
query_4_1 = f'''
-- для получения результата запроса необходимо объединить две таблицы БД: "areas" и "vacancies".
-- чтобы соотнести количество вакансий с конкретным регионом необходимо произвести объединение таблиц по ключу area_id

-- в качестве ответа на запрос получим названия регионов из столбца "name" таблицы "areas"
-- добавим к ним информацию о количестве вакансий в этом регионе, указанных в таблице "vacancies"
select 
    a.name as Город,
    count(v.id) as Число_вакансий
from vacancies as v
join areas as a on v.area_id = a.id

-- для корректной работы запроса необходимо сгруппировать данные по признаку "area_id"
group by 
    a.id
order by 2 DESC
'''
```

```python
# результат запроса
df_4_1 = pd.read_sql_query(query_4_1,connection)
print('Топ 10 регионов России по количеству вакансий')
df_4_1.head(10)
```

2. Напишите запрос, чтобы определить у какого количества вакансий заполнено хотя бы одно из двух полей с зарплатой.

```python
# текст запроса
query_4_2 = f'''
-- для получения необходимой информации воспользуемся таблице БД "vacancies"
-- выберем из таблицы количество вакансий для которых указана хотя бы одна из зарплатных вилок
-- для того чтобы ответ на SQL-запрос был более информативным, добавим общее количество вакансий и выведем результат в виде таблицы

-- в первой строке таблицы получим информацию о количестве в которых заполнено хотя бы одно из двух полей с зарплатой 
select 
    'cnt_not_null_salary' as type,
    count(*) as cnt
from vacancies as v

-- пропишем условие для отбора не пустых ячеек
where
    v.salary_from is not null or v.salary_to is not null

-- для объединения двух строк используем операцию UNION
union

-- вторая строка таблицы содержит информацию о общем количестве вакансий представленных в БД
select 
    'total_cnt',
    count(*) as cnt
from vacancies as v
join areas as a on v.area_id = a.id
'''
```

```python
# результат запроса
df_4_2 = pd.read_sql_query(query_4_2,connection)
print('В базе данных "vacancies"',df_4_2.iloc[0][1],'вакансии имеют информацию о верхнем или нижнем уровне зарплаты.')
print('Это составляет',round(100*df_4_2.iloc[0][1]/df_4_2.iloc[1][1]),'% от общего количества вакансий представленных в БД')
```

3. Найдите средние значения для нижней и верхней границы зарплатной вилки. Округлите значения до целого.

```python
# текст запроса

query_4_3 = f'''
-- для получения необходимой информации воспользуемся таблице БД "vacancies"
-- посчитаем среднее значение для каждой зарплатной вилки с помощью агрегирующей функции avg
-- для информативности ответа на sql-запрос добавим данные о наименьшем и наибольшем значение для зарплатной вилки
-- учтем, что пустые строки (Null) игнорируются агрегатными функциями и не будем фильтровать данные в таблице для их исключения

select
    avg(salary_from) as avg_salary_from,
    avg(salary_to) as avg_salary_to,
    max(salary_to) as max_salary_to,
    min(salary_from) as min_salary_from
from vacancies as v

'''
# при округление числа до целого с помощью операции round в sql для типа данных float остается 0 после запятой
# поэтому операцию округления до целого произведем при выводе результатов с помощью python
```

```python
# результат запроса
df_4_3 = pd.read_sql_query(query_4_3,connection)
print ('Среднее значение нижней границы зарплатной вилки составляет: ',round(df_4_3.iloc[0][0]), 'рублей.')
print ('Среднее значение верхней границы зарплатной вилки составляет: ',round(df_4_3.iloc[0][1]), 'рублей.')
print ('При этом минимальная нижняя граница зарплатной вилки составляет: ',round(df_4_3.iloc[0][3]), 'рублей.')
print ('Максимальная верхняя граница зарплатной вилки составляет: ',round(df_4_3.iloc[0][2]), 'рублей.')
```

4. Напишите запрос, который выведет количество вакансий для каждого сочетания типа рабочего графика (schedule) и типа трудоустройства (employment), используемого в вакансиях. Результат отсортируйте по убыванию количества.

5. Напишите запрос, выводящий значения поля Требуемый опыт работы (experience) в порядке возрастания количества вакансий, в которых указан данный вариант опыта. 

```python
# текст запроса
query_4_5 = f'''
-- для получения необходимой информации воспользуемся таблице БД "vacancies"
-- в качестве ответа на запрос получим данные ячеек из столбца "experience"
-- добавим к ним информацию о количестве вакансий в зависимости от опыта_работы
-- рассчитаем общее количество вакансий
with total as (
    select count(*) as cnt_vacancies
from vacancies
)

select 
    v.experience as Опыт_работы,
    count(v.id) as Число_вакансий,
    (100*count(v.id)/(select cnt_vacancies from total)) as процент -- добавим столбец с процентом от общего количества вакансий
from vacancies as v

-- для требуемых данных необходимо сгруппировать данные по признаку "experience"
group by 
    v.experience
order by 2
'''
```

```python
# результат запроса
df_4_5 = pd.read_sql_query(query_4_5,connection)
print('Распределение количества вакансий от необходимого опыта работы с процентом от общего количества вакансий')
df_4_5
```

# Юнит 5. Анализ работодателей

1. Напишите запрос, который позволит узнать, какие работодатели находятся на первом и пятом месте по количеству вакансий.

```python
# текст запроса
query_5_1 = f'''
-- для получения результата запроса необходимо объединить две таблицы БД: "employers" и "vacancies".
-- чтобы соотнести количество вакансий с конкретной компанией необходимо произвести объединение таблиц по ключу с id компаний

-- в качестве ответа на запрос получим имена компаний из ячеек столбца "name" таблицы "employers"
-- добавим к ним информацию о количестве вакансий этой компании, указанных в таблице "vacancies"
select
    e.name as компания, 
    count(v.id) as Число_вакансий
from vacancies as v

-- объединяем таблицы по ключу employer_id
join employers as e on v.employer_id = e.id

-- для требуемых данных необходимо сгруппировать данные по признаку "id" из таблицы "employers" 
group by    
    e.id

-- для ответа, ограничим запрос 5 первыми компаниями по количеству вакансий
order by 2 desc
limit 5
'''
```

```python
# результат запроса
df_5_1 = pd.read_sql_query(query_5_1,connection)
print('Топ 5 компаний по количеству предоставляемых вакансий')
print(df_5_1)
print()
print('Первое место в списке по количеству вакансий -', df_5_1.iloc[0][0])
print('Пятое место в списке по количеству вакансий -', df_5_1.iloc[4][0])
```

```python
# текст запроса
query_5_2_1 = f'''
-- для получения результата запроса необходимо объединить три таблицы БД: "areas", "vacancies" и "employers".
-- так как число регионов в таблице "areas" может оказаться больше, чем указано в таблицах "employers" и "vacancies",
-- то в случае объединения методом inner join мы потеряем часть регионов которые не представлены в таблицах "employers" и "vacancies"
-- с учетом того, что в основе анализа мы рассматриваем регионы, необходимо осуществить присоединение таблиц через операцию left join без потери информации о регионах
-- к таблице "areas" присоединим таблицу "employers" по ключу area_id
-- к таблице "areas" присоединим таблицу "vacancies" по ключу area_id

-- в качестве ответа на запрос получим названия регионов из ячеек столбца "name" таблицы "area"
-- добавим к ним информацию о количестве компаний зарегистрированных в этом регионе, указанных в таблице "employers"
-- добавим к ним информацию о количестве вакансий в этом регионе, указанных в таблице "vacancies"
-- учтем, что после соединения 3 таблиц количество ячеек размножится - 
-- например, в Батайске зарегистрировано 3 компании и размещено 8 вакансий. Результатом объединения будет 24 строки
-- для того, чтобы избежать дубликатов при расчете количества воспользуемся операцией DISTINCT  
select 
    a.name as Регион,
    count (distinct e.id) as число_компаний,
    count (distinct v.id) as число_вакансий
from areas as a

-- присоединяем таблицу "vacancies"
left join vacancies as v on v.area_id = a.id

-- присоединяем таблицу "employers"
left join employers as e on e.area = a.id

-- для требуемых данных необходимо сгруппировать данные по признаку "id" из таблицы "areas" 
group by 
    a.id

-- для анализа, ограничим запрос 20 первыми компаниями по количеству компаний
order by 2 desc
limit 10
'''
```

```python
df_5_2_1 = pd.read_sql_query(query_5_2_1,connection)
print('Топ 10 регионов по количеству зарегистрированных компаний и число вакансий в этих регионах')
print(df_5_2_1)
```

```python
# результат запроса
df_5_2_2 =pd.read_sql_query(query_5_2_2,connection)
print('На первом месте по числу зарегистрированных компаний среди регионов, о которых нет информации в таблице "vacancies" находится', df_5_2_2.iloc[0][0]+'.')
print('В этом регионе зарегистрировано',df_5_2_2.iloc[0][1], 'вакансий.')
```

Если возникнут трудности с этим задание посмотрите материалы модуля  PYTHON-17. Как получать данные из веб-источников и API. 

```python
# код для получения списка городов-милионников
import requests # импорт библиотеки для работы с веб-запросами
from bs4 import BeautifulSoup #импорт библиотеки для работы с html страницами
url = 'https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8'
response = requests.get(url) # Отправляем запрос на получение данных с веб-страницы
page = BeautifulSoup(response.text,'html.parser') # Преобразуем полученные данные в html-текст
table_cites = page.find('table', class_="standard sortable") # При помощи необходимого хештега ищем таблицу с городами-миллиониками

# Преобразуем полученную информацию в DataFrame
headers =[] # Создаем список будущих заголовков таблицы
# Заполняем список заголовками таблицы с помощью цикла 
for th in table_cites.find_all('th'):
    headers.append(th.text)
# По полученным заголовка формируем заготовку под DataFrame
my_table_cites = pd.DataFrame(columns = headers)
# Заполняем DataFrame данными
for tr in table_cites.find_all('tr')[1:]: # Внутри хештега <tr> - находится данные строки таблицы
    row = [td.text for td in tr.find_all('td')] # Внутри хештега <td> - находится данные ячеек таблицы. Заполняем строку данными из ячеек. 
    length = len(my_table_cites)
    my_table_cites.loc[length] = row
list_cites = my_table_cites['Город'].to_list() #получим список городов-милионников
```

```python
# Создадим строку (шаблон) со списком городов-милионников  и условием для фильтрации sql-запроса [areas.name in ("Название города 1","Название города 2",...)]
str_cites = 'a.name in ('
for i in list_cites:
    str_cites+="'"+i+"'"+","
str_cites = str_cites[:-1]+')' # добавляем скобку в конец и убираем лишнюю запятую

# Формируем текст запроса

# Начало запроса
# для получения результата запроса необходимо объединить три таблицы БД: "employers", "vacancies" и "areas".
# так как нам нужно получить подробную информацию только о компании "Яндекс", то можно использовать метод объединения inner join
# к таблице "employers" присоединим таблицу "vacancies" по ключу employer_id
# к таблице "vacancies" присоединим таблицу "areas" по ключу area_id
# по заданию, добавим строку Total с общим количеством вакансий компании в этих регионах
query_5_7_0 =f'''
select
    e.name as Компания,
    a.name as Город,
    count(a.name) as Число_вакансий
from employers as e
join vacancies as v on e.id = v.employer_id
join areas as a on v.area_id = a.id
where e.name = 'Яндекс' and
'''
# продолжение 1
query_5_7_1 = f'''
group by 
    e.name,
    a.name
union
select
    e.name,
    'total' as total,
    count(a.id)
from employers as e
join vacancies as v on e.id = v.employer_id
join areas as a on v.area_id = a.id
where e.name = 'Яндекс' and
'''
# продолжение 2
query_5_7_2 = f'''
group by 
    e.name
order by 3 desc
'''

# текст запроса
query_5_7 = query_5_7_0 + str_cites + query_5_7_1 + str_cites + query_5_7_2
```

* в названии есть слово 'junior' *или*
* требуемый опыт — Нет опыта *или*
* тип трудоустройства — Стажировка.
 

```python
# текст запроса
query_6_2 = f'''
-- для получения необходимой информации воспользуемся таблице БД "vacancies"
-- в качестве ответа на запрос посчитаем количество строк таблицы при соблюдение условий задания 
-- для реализации воспользуемся оператором like и значком %(искомые слова могу находится в тексте на любой позиции)
-- обратим внимание на то, что для поиска вакансий с названием включающем 'ML'(Machine Learning) необходимо сохранить регистр букв без изменений
-- в противном случае мы можем получить ложные значение, такие например как 'teaMLead'
-- так же обратим внимание, что необходимо "очистить" данные от таких вакансий которые содержат 'C++ разработчик(QML)'
select 
    'DS-junior' as type,
    count(*) as Число_вакансий
from vacancies as v
where 
    (lower(name) like '%data scientist%' 
    or lower(name) like '%data science%' 
    or lower(name) like '%исследователь данных%'
    or (name like '%ML%' and lower(name) not like '%html%' and lower(name) not like '%qml%')
    or lower(name) like '%machine learning%'
    or lower(name) like '%машинн%обучен%%'
    )
and
    (lower(name) like '%junior%' 
    or experience = 'Нет опыта'
    or employment = 'Стажировка'
    )
union
select 
    'Общее число вакансий',
    count(*)
from vacancies as v
'''
```

```python
# результат запроса
df_6_2 = pd.read_sql_query(query_6_2,connection)
print('Для начинающего Data science специалиста в БД представлено',df_6_2.iloc[0][1],'вакансия.')
print('Это составляет',round(100*df_6_2.iloc[0][1]/df_6_2.iloc[1][1],2),'% от общего числа вакансий.')
```

Ответ округлите до двух знаков после точки-разделителя.

```python
# текст запроса
query_6_5 = f'''
-- для получения необходимой информации воспользуемся таблице БД "vacancies"
-- обратим вниманием, что ключевые навыки разделены знаком табуляции
-- поэтому, для подсчета числа ключевых навыков в конкретной вакансии нужно посчитать число знаков табуляции
-- число ключевых навык на единицу больше числа знаков табуляции
select 
    round(avg(length(key_skills) - length(replace(key_skills,chr(9),'')) + 1),2)
from vacancies as v
where 
    (
    lower(name) like '%data scientist%' 
    or lower(name) like '%data science%' 
    or lower(name) like '%исследователь данных%'
    or (name like '%ML%' and lower(name) not like '%html%' )
    or lower(name) like '%machine learning%'
    or lower(name) like '%машинн%обучен%'
    )
and key_skills is not null
'''
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
# текст запроса
query_6_6 = f'''
-- для получения необходимой информации воспользуемся таблице БД "vacancies"
-- обратим внимание на то что в задание указано нужно найти среднюю зарплату для двух случаев
-- для реализации этого условия задания воспользуемся оператором case..when..then..else..end 
select
    experience as Опыт_работы,
    round(
    avg(
    case
        when salary_from is not null and salary_to is not null -- если заполнены оба полей зарплатной вилки
        then (salary_to+salary_from)/2 -- рассчитываем ЗП в случае выполнения условия
        else (coalesce(salary_to,0)+coalesce(salary_from,0)) -- рассчитываем ЗП в случае если условие не выполняется (учитываем, что есть условие на всю таблицу)
    end
    )) as Средняя_зарплата  
from vacancies as v
where 
    (
    lower(name) like '%data scientist%' 
    or lower(name) like '%data science%' 
    or lower(name) like '%исследователь данных%'
    or (name like '%ML%' and lower(name) not like '%html%' )
    or lower(name) like '%machine learning%'
    or lower(name) like '%машинн%обучен%'
    )
and 
    (salary_from is not null or salary_to is not null)
group by 
    experience
order by 2 desc
'''
```

```python
# результат запроса
df_6_6 = pd.read_sql_query(query_6_6,connection)
print('Распределение средней заработной платы Data Science специалиста в зависимости от опыта работы')
print(df_6_6)
print()
print('C опытом работ от 3 до 6 лет Data Science специалист может рассчитывать на среднюю заработную плату равную',round(df_6_6[df_6_6['Опыт_работы']=='От 3 до 6 лет'].iloc[0][1]),'рублей.')
```
